```python
import os
import numpy as np
from PIL import Image

def convert_npy_to_png(source_folder, dest_folder):
    # Iterate through subfolders in the source folder
    for subdir, _, files in os.walk(source_folder):
        for file in files:
            if file.endswith(".npy"):
                # Load the npy file
                data = np.load(os.path.join(subdir, file))
                
                # Check if the shape is (22, 160, 240)
                if data.shape == (22, 160, 240):
                    # Get the subfolder name and create it in the destination folder
                    subfolder_name = os.path.basename(subdir)
                    new_subfolder = os.path.join(dest_folder, subfolder_name)
                    os.makedirs(new_subfolder, exist_ok=True)

                    # Save images as png files in the destination subfolder
                    for idx, img in enumerate(data):
                        img_path = os.path.join(new_subfolder, f"image_{idx}.png")
                        img_pil = Image.fromarray(img.astype(np.uint8))
                        img_pil.save(img_path)

# Set your source folder and destination folder paths
source_folder = "Dataset_Student/train/"
dest_folder = "Dataset_Student/train_targets/"

# Call the function to convert npy files to png images
convert_npy_to_png(source_folder, dest_folder)
```

```python
import os
import shutil

def flatten_directory(source_folder, dest_folder):
    # Iterate through subfolders in the source folder
    for subdir, _, files in os.walk(source_folder):
        for file in files:
            if file.endswith(".png"):
                # Get the subfolder name and create the destination folder if it doesn't exist
                subfolder_name = os.path.basename(subdir)
                os.makedirs(dest_folder, exist_ok=True)

                # Create the new file name by joining the subfolder name and the file name
                new_file_name = f"{subfolder_name}_{file}"

                # Copy the file to the destination folder with the new file name
                src_file_path = os.path.join(subdir, file)
                dest_file_path = os.path.join(dest_folder, new_file_name)
                shutil.copy(src_file_path, dest_file_path)

# Set your source folder and destination folder paths
source_folder = "Dataset_Student/train/"
dest_folder = "Dataset_Student/train_flat/"

# Call the function to flatten the directory and copy the png files
flatten_directory(source_folder, dest_folder)
```
