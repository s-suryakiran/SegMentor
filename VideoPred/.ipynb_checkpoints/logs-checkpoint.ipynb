```python
logfile = open("mainlog.txt","r")
mse_list=[]
mae_list=[]
val_loss_list=[]
train_loss_list=[]
lr_list=[]

for i,line in enumerate(logfile):
    if(line.startswith("val")):
        mse,mae = line.split(",")
        mse = mse.strip()
        mae = mae.strip()
        mse=mse[9:]
        mae=mae[4:]
        mse_list.append(float(mse))
        mae_list.append(float(mae))
    if(line.startswith("Epoch")):
        steps,lr,train_loss,vali_loss = line.split("|")
        lr = lr.strip()
        steps = steps.strip()
        train_loss=train_loss.strip()
        vali_loss=vali_loss.strip()
        lr = lr[4:]
        train_loss=train_loss[12:]
        vali_loss = vali_loss[11:]
        lr_list.append(float(lr))
        train_loss_list.append(float(train_loss))
        val_loss_list.append(float(vali_loss))

epochs = list(range(1,len(val_loss_list)+1))

```

```python
import matplotlib.pyplot as plt
plt.plot(epochs,mse_list)
plt.xlabel("Epochs")
plt.ylabel("MSE")
plt.savefig("./images/MSE.png")
plt.show()
```

```python
plt.plot(epochs,mae_list)
plt.xlabel("Epochs")
plt.ylabel("MAE")
plt.savefig("./images/MAE.png")
plt.show()
```

```python
plt.plot(epochs,val_loss_list)
plt.xlabel("Epochs")
plt.ylabel("Val. Loss")
plt.savefig("./images/valloss.png")
plt.show()
```

```python
plt.plot(epochs,train_loss_list)
plt.xlabel("Epochs")
plt.ylabel("Train Loss")
plt.savefig("./images/trainloss.png")
plt.show()
```

```python
plt.plot(epochs,lr_list)
plt.xlabel("Epochs")
plt.ylabel("LR")
plt.savefig("./images/lr.png")
plt.show()
```
